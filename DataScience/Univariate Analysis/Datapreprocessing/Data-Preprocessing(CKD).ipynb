```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual


    
```

```python
quan,qaul=quanQual(dataset)
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Define the imputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fit the imputer on the numerical columns (quan) of the dataset
imp.fit(dataset[quan])

# Transform the dataset by replacing missing values with the mean
df = imp.transform(dataset[quan])

# Convert the resulting array back into a DataFrame with the same column names
df = pd.DataFrame(df, columns=quan)
```

```python
cate=dataset[qaul]
```

```python
two=[df,cate]
```

```python
preprocessed=pd.concat(two,axis=1)
```
